# Part 1: Cardiovascular Disease Classifier
Using the "cardio_train.csv" to train an XGBoost classifier to predict whether a patient has cardivascular disease or not.

```python
# import the libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# read the csv file 
df = pd.read_csv("cardio_train.csv", sep=";")
```

```python
df.head(3)
```

```python
# Drop id
df = df.drop(columns = 'id')
```

```python
# since the age is given in days, we convert it into years
df['age'] = df['age']/365
```

```python
df.head(3)
```

```python
# Statistical summary of the dataframe
df.describe()
```

```python
df.hist(bins = 30, figsize = (20,20), color = 'r')
plt.show()
```

```python
# get the correlation matrix
corr_matrix = df.corr()
corr_matrix
```

```python
# plotting the correlation matrix
plt.figure(figsize = (16,16))
# annot True includes the correlation values 
sns.heatmap(corr_matrix, annot = True)
plt.show()
```

# Split into Train and Test datasets

```python
# split the dataframe into target and features
y = df['cardio'] #output
X = df.drop(columns =['cardio']) #input
```

```python
X.shape
```

```python
y.shape
```

```python
#spliting the data in to test and train sets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
X_test.shape
```

```python
y_test.shape
```

# Train an XGBoost Classifier

```python
!pip install xgboost
```

```python
from xgboost import XGBClassifier
```

```python
# Train an XGBoost classifier model 
# predicting whether patient has cardivascular disease or not (binary)
xgb_classifier = XGBClassifier(objective ='binary:logistic', eval_metric = 'error', learning_rate = 0.1, max_depth = 1, n_estimators = 10)
xgb_classifier.fit(X_train, y_train)
```

# Test XGBoost Classifier

```python
# predict the score of the trained model using the testing dataset
result = xgb_classifier.score(X_test, y_test)
print("Accuracy : {}".format(result))
```

```python
# make predictions on the test data
# 0 no disease, 1 disease
y_predict = xgb_classifier.predict(X_test)
y_predict
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_predict))
# precision TP/(TP+FP)
# recall TP/(TP+FN)
# F-beta score can be interpreted as a weighted harmonic mean of the precision and recall
# where an F-beta score reaches its best value at 1 and worst score at 0. 
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_predict)
sns.heatmap(cm, fmt = 'd', annot = True)
```

```python
# see if a higher max_depth improves accuracy
xgb_classifier_deep = XGBClassifier(objective ='binary:logistic', eval_metric = 'error', learning_rate = 0.1, max_depth = 6, n_estimators = 10)
xgb_classifier_deep.fit(X_train, y_train)
result_deep = xgb_classifier_deep.score(X_test, y_test)
print("Accuracy : {}".format(result_deep))
```

# Part 2: Diabetes Classifier
Using the "diabetes.csv" to train an XGBoost classifier to predict whether a patient has diabetes or not.

```python
df = pd.read_csv('diabetes.csv')
df.head()
```

```python
df.info()
```

```python
# Plot Histogram
df.hist(bins = 30, figsize = (20,20), color = 'r');
```

```python
# Plot the correlation matrix
correlations = df.corr()
f, ax = plt.subplots(figsize = (10, 10))
sns.heatmap(correlations, annot = True);
```

# Split into Train and Test datasets

```python
y = df['Outcome']
X = df.drop(['Outcome'], axis = 1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
# Train an XGBoost classifier model 
xgb_classifier = XGBClassifier(objective ='binary:logistic', eval_metric = 'error', learning_rate = 0.1, max_depth = 1, n_estimators = 10)
xgb_classifier.fit(X_train, y_train)
```

```python
# predict the score of the trained model using the testing dataset
result = xgb_classifier.score(X_test, y_test)
print("Accuracy : {}".format(result))
```

```python
# make predictions on the test data
y_predict = xgb_classifier.predict(X_test)
y_predict
```

```python
print(classification_report(y_test, y_predict))
```

```python
cm = confusion_matrix(y_test, y_predict)
sns.heatmap(cm, fmt = 'd', annot = True)
```
